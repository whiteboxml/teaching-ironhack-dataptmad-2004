# Pivot Tables

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html

```python
import numpy as np
import pandas as pd

pd.options.display.max_columns = None
```

## 1. Read dataset

```python
vehicles = pd.read_csv('../../week_05/pandas_import_export/data/vehicles.csv')
vehicles.head()
```

## 2. Aggregations

```python
# look for distinct values and their occurrences
vehicles['Vehicle Class'].value_counts()
```

```python
# the same but using group by
vehicles.groupby('Vehicle Class')['Vehicle Class'].agg('count').sort_values(ascending=False)
vehicles.groupby('Vehicle Class')['Vehicle Class'].count().sort_values(ascending=False)
vehicles.groupby('Vehicle Class')['Vehicle Class'].agg(len).sort_values(ascending=False)
vehicles.groupby('Vehicle Class').agg({'Vehicle Class': 'count'})['Vehicle Class'].sort_values(ascending=False)
```

```python
# aggregate other columns by ??? for each type of vehicle class
vehicles.pivot_table(index=['Vehicle Class'])
```

```python
# check that the aggregation function is the mean
vehicles[vehicles['Vehicle Class'] == 'Vans Passenger']
```

```python
# the same but using group by
vehicles.groupby('Vehicle Class').mean()
```

```python
# create multicolumn dataframe
vehicles.pivot_table(index=['Vehicle Class'], columns=['Cylinders'])
```

```python
vehicles[(vehicles['Vehicle Class'] == 'Large Cars') & (vehicles['Cylinders'] == 5.0)]
```

```python
vehicles.pivot_table(index=['Vehicle Class'], columns=['Cylinders', 'Engine Displacement'])
```

```python
complex_df = vehicles.pivot_table(index=['Vehicle Class'], columns=['Cylinders'])

complex_df.columns
```

```python
# select which columns to aggregate and the aggregation function
vehicles.pivot_table(index=['Vehicle Class'], values=['Combined MPG'])
```

```python
vehicles.pivot_table(index=['Vehicle Class'], values=['Combined MPG'], aggfunc=len)['Combined MPG'].sort_values(ascending=False)
```

```python
df_agg = vehicles.pivot_table(index=['Vehicle Class'], columns=['Cylinders'], values=['Combined MPG'], fill_value=0.0)

df_agg
```
